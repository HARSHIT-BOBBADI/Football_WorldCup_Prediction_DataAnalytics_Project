```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
years = [1930,1934,1938,1942,1946,1950,1954,1958,1962,1966,1970,1974,1978,1982,1986,1990,1994,1998,2002,2006,2010,2014,2018]
```

```python
def get_matches(year):
    web = f'https://en.wikipedia.org/wiki/{year}_FIFA_World_Cup'
    response = requests.get(web)
    content = response.text

    soup = BeautifulSoup(content, 'lxml')
    matches = soup.find_all('div', class_="footballbox")

    home = []
    away = []
    score = []

    for match in matches:
      home.append(match.find('th', class_='fhome').get_text())
      score.append(match.find('th', class_='fscore').get_text())
      away.append(match.find('th', class_='faway').get_text())

    dict_football = {'home' : home, 'score' : score, 'away' : away}
    df_football = pd.DataFrame(dict_football)
    df_football['year'] = year

    return df_football
```

```python
# historical data

fifa = [get_matches(year) for year in years]
df_fifa = pd.concat(fifa, ignore_index = True)
df_fifa.to_csv('fifa_worldcup_historical_data.csv', index=False)
```

```python
# another function because the data in the website has changed post worldcup

def get_fixture_matches(url):
    response = requests.get(url)
    content = response.text

    soup = BeautifulSoup(content, 'lxml')
    matches = soup.find_all('div', class_="footballbox")

    home = []
    away = []
    score = []

    for match in matches:
        home.append(match.find('th', class_='fhome').get_text())
        away.append(match.find('th', class_='faway').get_text())
        score.append(match.find('th', class_='fscore').get_text())

    dict_fixtures = {'home': home, 'score': score, 'away': away}
    df_fixtures = pd.DataFrame(dict_fixtures)

    return df_fixtures
```

```python
# fixture data

url_fixtures = "https://web.archive.org/web/20221115040351/https://en.wikipedia.org/wiki/2022_FIFA_World_Cup"
df_fixtures = get_fixture_matches(url_fixtures)
df_fixtures.to_csv('fifa_worldcup_2022_fixtures.csv', index=False)
```
